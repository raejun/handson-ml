```python
# 파이썬 그래프 한글 깨지지않게 설정하는 코드
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import platform
import matplotlib

if platform.system() == 'Windows': # 윈도우인 경우
    font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)  # 'Malgun Gothic'
else: # Mac 인 경우, AppleGothic(기본 글꼴)
    rc('font', family='AppleGothic')
    # 주의 : 맥에서는 나눔글꼴이름이 NanumGothic 이 아니라 NanumGothicOTF 임

matplotlib.rcParams['axes.unicode_minus'] = False #그래프에서 마이너스 기호가 표시되도록 하는 설정
```

```python
import numpy as np
data = np.random.randint(-100, 100, 50).cumsum()

plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.style.use('seaborn-pastel')
plt.show()
```

```python
# OSX의 설치된 폰트를 가져오는 함수
from matplotlib import font_manager
font_list_mac = font_manager.OSXInstalledFonts()
print(len(font_list_mac))
# font_list_mac
sorted(font_list_mac)
```

```python
from matplotlib import font_manager
fonts = [f.name for f in font_manager.fontManager.ttflist if '' in f.name]
sorted(fonts)
```

```python
from matplotlib import font_manager
nanum_fonts = [(f.name, f.fname) for f in font_manager.fontManager.ttflist if 'Nanum' in f.name]
sorted(nanum_fonts)
```
